```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Read the Dataset

df = pd.read_csv("V:/7 Mentors/Amazon Sale Report.csv")
df
```

```python
# along with finding whether they contain null values or not.

df.info()
```

```python
#Finding how many null values in dataset
df.isnull().sum()
```

```python
df.columns
```

```python
#removing unnecessary columns 
df = df.drop('index' , axis = 1)
```

```python
df.head(3)
```

```python
#Removing Duplicates
df = (df.drop_duplicates(keep='first').reset_index(drop=True)) 
```

```python
df.isna().sum()
```

```python
# Fill the null values with mode and median

null_coll = ['Courier Status', 'currency', 'ship-city', 'ship-state', 'ship-country']
for i in null_coll:
    mode = df[i].mode()[0]
    df[i] = df[i].fillna(mode)
df['Amount'] = df['Amount'].fillna(df['Amount'].median())
```

```python
df.isna().sum()
```

# Handling The Outliers

```python
# Visualize the distribution of numerical features before handling outliers
plt.figure(figsize =(20,30))
sns.boxplot(data=df)
plt.title("Boxplot before handling outliers")
plt.show()
```

```python
Q1 = df['ship-postal-code'].quantile(0.25)
Q3 = df['ship-postal-code'].quantile(0.75)
IQR = Q3 - Q1
print("Q1 : ", Q1)
print("Q3 : ", Q3)
print("IQR Value : ", IQR)
lower_range = Q1 - 1.5*IQR
upper_range = Q3 + 1.5 *IQR

# Identify outliers
outliers = df[(df['ship-postal-code'] < lower_range) | (df['ship-postal-code'] > upper_range)]
print("\nOutliers:- \n", outliers)

#Remove outliers
df_no_outliers = df[(df['ship-postal-code'] >= lower_range) | (df['ship-postal-code'] <= upper_range)]

print("\.DataFrame after removing outliers:")
print(df_no_outliers.shape)
```

```python
df.isna().sum()
```

### **1.Distribution Of Sales Amount Across Count Of Orders**

```python
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1
upper_range = Q3 + 1.5 * IQR
lower_range = Q1 - 1.5 * IQR
Median = df['Amount'].median()
df['Amount'] = df['Amount'].apply(lambda x: Median if x > upper_range  or x < lower_range else x) 
```

```python
df['Amount'] = df['Amount'].apply(lambda x: Median if x == 0 else x) 
```

```python
fig, S= plt.subplots(figsize=(8,5))
sns.histplot(data = df['Amount'],ax = S, binwidth = 100, kde = True)
plt.grid(linestyle= '--', color = '#000000')
plt.title('Distribution Of Sales Amount Across Count Of Orders')
plt.tight_layout()
```

### Here we can see that the most of the sales are between 400rs to 600rs

### **2.Now we can see that the top 5 trending products**

```python
Top5_product=df.groupby('Category')['Amount'].sum().to_frame().sort_values(by=['Amount'],ascending=False).head(5)
Top5_product
```

```python
Top5_product.plot(kind='bar',color='#f0bda7')
plt.title('Total Amount by Category', fontsize=18, weight='bold')
plt.ylabel('Total Amount (in millions)', fontsize=14)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```

**Here we can see that the total amount of sales by category**
